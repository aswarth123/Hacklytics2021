```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data = pd.read_csv("~/raw_data/testing.csv")
data.head()
```

```python
data.loc[data['positivity_rate']>=100]
```

```python
#removed the outliers
data = data.loc[data['positivity_rate']<=100]
```

```python
#testing_rate has no outliers
len(data.loc[round(data['tests_done']/data['population']*100000,6) == round(data['testing_rate'],6)])
```

```python
#removed the regions(subnational) and kept only the countries
national_data = data.loc[data['level'] == 'national']
data.head()
```

```python
national_data.drop(['country_code', 'level', 'region', 'region_name', 'testing_data_source'], axis = 1, inplace = True)
national_data.head()
```

```python
occ_data = pd.read_csv("~/raw_data/data.csv")
occ_data.head()
```

```python
#dropped unnecessary columns
occ_data.drop(['source', 'url'], axis = 1, inplace = True)
occ_data.head()
```

```python
#dropped the rows which had empty dates
occ_data = occ_data.dropna()
```

```python
#merged hospital_occupancy and testing datasets on country and year_weak
merged_dataframe = pd.merge(national_data,occ_data,on=['country','year_week'])
```

```python
kaggle_data = pd.read_csv("~/raw_data/kaggle_data.csv")
kaggle_data.head()
```

```python
kaggle_data.drop(['CODE'], axis = 1, inplace = True)
kaggle_data.head()
```

```python
#merged the data with the kaggle dataset on the basis of country and date
final_data = pd.merge(merged_dataframe, kaggle_data, on=['country','date'])
```
